**Clean up dataset**

```python
random_sen = pd.read_csv("Random_Sen_2000.csv")
```

```python
#removing duplicates
random_sen = random_sen.drop_duplicates()
```

```python
len(random_sen)
```

```python
#depression keywords taken from DSM-5, BDI-II, and synonyms (taken from thesaurus.com)
keywords = ["depressed", "depression", "loss of interest", "loss of pleasure", "sad", "empty", "hopeless", "irritable",
           "diminished interest", "decreased interest", "diminished pleasure", "decreased pleasure", "weight loss", 
            "weight gain", "decrease in appetite", "increase in appetite", "insomnia", "hypersomnia", "fatigue", "tired", 
            "loss of energy", "worthlessness", "guilt", "indecisiveness", "indecision", "death", "suicide", "die", "kill", 
            "failure", "punishment", "self-dislike", "self-hate", "cry", "agitation", "agitated", "social withdrawal", 
            "difficulty concentrating", "loss of interest in sex", "fatigued", "listless", "listlessness", "exhausted", "exhaustion", "breakdown", "failing", "penalty", "punished",
            "punish", "suffering", "suffer", "torture", "worthless", "useless", "inadequate", "inadequacy", "uselessness",
            "regret", "shame", "sin", "remorse", "regretful", "shameful", "remorseful", "indecisive", "hesitation", "hesitate",
            "uncertainty", "uncertain", "doubt", "doubtful", "dying", "self-hatred", "unrest", "restless", "turmoil", 
             "loneliness", "lonely", "alone"]
```

```python
#checking sentences containing the keywords
for word in keywords:
    for sentence in random_sen['Sentence']:
        if word in sentence:
            print("Sentence: ")
            print(sentence)
            print("\n")
```

```python
#finding user of sentences indicating depression of the user (two found)

sentences =[
"I've seen some shit and most days I wish it was me that died instead of her...",
"I’m on a couple medications for my bipolar and I feel guilty about it but they’ve said they’re relatively safe and it’s be worse for me to come off them."
]

users = []
for i in range(len(random_sen)):
    for sentence in sentences:
        #print(sentence)
        if random_sen.iloc[i]["Sentence"] == sentence:
            users.append(random_sen.iloc[i]["User"])
        
print(users)
        
#removing the user
for user in users:
    random_sen = random_sen[random_sen['User'] != user]

#checking that user and their sentences have been removed
len(random_sen)
len(random_sen["User"].unique())
```

```python
random_sen = random_sen.reset_index(drop=True)
```

```python
#removing sentences that are shorter than 2 words (to avoid removed sentences, sentences of only an emoji or exclamation mark etc.)
drop_list = []
for i in range(len(random_sen)):
    #print(i)
    sentence = random_sen.iloc[i]["Sentence"]
    words = sentence.split()
    if len(words) < 2:
        drop_list.append(i)
    
len(drop_list)

random_sen = random_sen.drop(drop_list)
random_sen = random_sen.reset_index(drop=True)
```

**Retrieving a subset of sentences from male and female users equal to the depressed dataset**

```python
#retrieving a subset of 9 female users
female_users = random_sen[random_sen['Gender'] == 1]['User'].unique()
subset_female_users = pd.Series(female_users).sample(n=9, replace=False)
subsetfemale_df = random_sen[random_sen['User'].isin(subset_female_users)]
```

```python
subsetfemale_df.head()
```

```python
len(subsetfemale_df)
```

```python
#sampling 337 sentences
control_female_sen = subsetfemale_df.sample(n=337)
control_female_sen.to_csv("Control_Female_Sen.csv", index=False)
```

```python
#retrieving a subset of 11 male users
male_users = random_sen[random_sen['Gender'] == 0]['User'].unique()
subset_male_users = pd.Series(male_users).sample(n=11, replace=False)
subsetmale_df = random_sen[random_sen['User'].isin(subset_male_users)]
```

```python
subsetmale_df.head()
```

```python
len(subsetmale_df)
```

```python
#sampling 212 sentences
control_male_sen = subsetmale_df.sample(n=212)
control_male_sen.to_csv("Control_Male_Sen.csv", index=False)
```

```python
#concatenating the male and female sentences
control_sen = pd.concat([control_male_sen, control_female_sen], ignore_index=True)
#control_sen.to_csv("Gendered_Control_Sen.csv", index=False)
```

**Bringing the control sentences dataset into the same format as the depressed dataset**

```python
found_bdisen = pd.read_csv("Depressed-Found-BDI-Sen-Users-Genders.csv")
```

```python
#renaming the user column as subject as that is used in the original dataset
control_sen = control_sen.rename(columns={'User': 'Subject'})
```

```python
#adding the columns for symptoms and assigning them the right value
for column in found_bdisen.columns:
    if column not in control_sen.columns:
        #print(column)
        control_sen[column] = 0

#severity of 4 is used for non-depressed users
control_sen["Severity"] = 4
```
